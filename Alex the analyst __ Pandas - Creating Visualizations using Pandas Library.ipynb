## Reading in dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read the csv
df = pd.read_csv(r'C:\Users\HP\Downloads\Ice Cream Ratings.csv')

# show the dataframe
df
```

```python
# set the column Date as index
df = df.set_index('Date')

# show the dataframe
df
```

## Plot() in Pandas

```python
df.plot()
```

```python
df.plot(kind = 'line')
```

```python
# Plotting multiple subplots for each column in the DataFrame  
df.plot(subplots=True)
```

```python
# Not plotting multiple subplots for each column in the DataFrame 
df.plot(subplots = False)
```

```python
# Plotting the line graph for ice cream ratings with specified title and axis labels  
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

### Bar plot

```python
# Plotting the bar chart for df
df.plot(kind = 'bar')
```

```python
# Plotting the stacked bar chart for df
df.plot(kind = 'bar', stacked = True)
```

```python
# Plotting the stacked bar chart for 'Flavor Rating'
df['Flavor Rating'].plot(kind = 'bar', stacked = True)
```

```python
# Plotting the horizontal stacked bar chart for df
df.plot.barh(stacked = True)
```

### Scatter plot    

```python
# Plotting a scatter plot and specifying the axis labels
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating')
```

```python
# Plotting a scatter plot and specifying the axis labels and the size of each point (100) 
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 100)
```

```python
# Plotting a scatter plot and specifying the axis labels and the size of each point (500) 
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500)
```

```python
# Plotting a scatter plot and specifying the axis labels, the size of each point (500) and the color (Yellow) 
df.plot.scatter(x = 'Texture Rating', y = 'Overall Rating', s = 500, c = 'Yellow')
```

### Histogram

```python
# Plotting a histogram for df
df.plot.hist()
```

```python
# A histogram with bins = 10
df.plot.hist(bins = 10)
```

```python
# A histogram with bins = 20
df.plot.hist(bins = 20)
```

### Boxplot

```python
# A boxplot
df.boxplot()
```

### Area plot

```python
# An area plot
df.plot.area()
```

```python
# Plotting an area plot and specifying the size of the figure
df.plot.area(figsize = (10,5))
```

### Pie chart

```python
# Pie plot for 'Flavor Rating'
df.plot.pie(y = 'Flavor Rating')
```

```python
# specifying the size of the figure
df.plot.pie(y = 'Flavor Rating', figsize = (10,6))
```

## Choosing styles

```python
# Creating a line plot for Ice Cream Ratings, labeling the axes and providing a title  
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
# Displaying the list of available styles in Matplotlib for customizing plots  
print(plt.style.available)
```

```python
# Setting the plot style to 'seaborn-v0_8-deep' for better aesthetics  
plt.style.use('seaborn-v0_8-deep')  

# Creating a line plot for Ice Cream Ratings, labeling the axes and providing a title  
df.plot(kind='line', title='Ice Cream Ratings', xlabel='Daily Ratings', ylabel='Scores')
```

```python
# Setting the plot style to 'fivethirtyeight' for better aesthetics
plt.style.use('fivethirtyeight')

# Creating a line plot for Ice Cream Ratings, labeling the axes and providing a title
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```

```python
# Setting the plot style to 'classic' for better aesthetics
plt.style.use('classic')

# Creating a line plot for Ice Cream Ratings, labeling the axes and providing a title
df.plot(kind = 'line', title = 'Ice Cream Ratings', xlabel = 'Daily Ratings', ylabel = 'Scores')
```
