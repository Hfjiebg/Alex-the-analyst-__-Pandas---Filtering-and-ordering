### Grouping data

```python
# Show df
df
```

- Are there certain continents that have grown faster than others, and in which ways?

```python
# Group the DataFrame by 'Continent' and calculate the mean for numeric columns  
df.groupby('Continent').mean(numeric_only=True)
```

```python
# Show the rows where the column 'Continent' contains 'Oceania'
df['Continent'].str.contains('Oceania')
```

```python
# Filter the DataFrame to include only rows where the column 'Continent' contains 'Oceania'  
df[df['Continent'].str.contains('Oceania')]
```

```python
# Group the DataFrame by 'Continent', calculate the mean for numeric columns,  
# and sort the results by '2022 Population' in descending order  
df.groupby('Continent').mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)
```

```python
# Define df2
df2 = df.groupby('Continent').mean(numeric_only=True).sort_values(by = '2022 Population', ascending = False)

# show df2
df2
```

```python
# Plot df2
df2.plot()
```

```python
# Tranpose the data & name it df3
df3 = df2.transpose()

# Show df3
df3
```

```python
# Plot df3
df3.plot()
```

```python
# Show df2
df2
```

```python
# Group the DataFrame by 'Continent', calculate the mean for specific columns (from index 5 to 12),  
# and sort the results by '2022 Population' in descending order  
df2 = df.groupby('Continent')[df.columns[5:13]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)  

# Display the resulting DataFrame  
df2
```

```python
# Group the DataFrame by 'Continent', calculate the mean for the specified columns (in reverse order),  
# and sort the results by '2022 Population' in descending order  
df2 = df.groupby('Continent')[df.columns[5:13][::-1]].mean(numeric_only=True).sort_values(by='2022 Population', ascending=False)  

# Display the resulting DataFrame
df2
```

```python
# Transpose the data & name it df3
df3 = df2.transpose()

# Show df3
df3
```

### Visualizing grouped data

```python
# Plot df3
df3.plot()
```

The answer : Asia grew faster than the other continents especially in the 2000s.
