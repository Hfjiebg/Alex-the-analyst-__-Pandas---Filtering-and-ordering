## Removing duplicates

```python
# removing the duplicates
df = df.drop_duplicates()

# show the dataframe
df
```

## Dropping columns

```python
# drop the column 'Not_Useful_Column'
df = df.drop(columns = 'Not_Useful_Column')

# show the dataframe
df
```

## Strip

```python
# Remove leading and trailing whitespace from the 'Last_Name' column  
df['Last_Name'] = df['Last_Name'].str.strip()
```

```python
# Remove leading occurrences of the substring '...' from the 'Last_Name' column  
df['Last_Name'] = df['Last_Name'].str.lstrip('...')

# Remove leading occurrences of the substring '/' from the 'Last_Name' column  
df['Last_Name'] = df['Last_Name'].str.lstrip('/')
```

```python
# Remove occurrences of the substring '_' from the left of the 'Last_Name' column
df['Last_Name'] = df['Last_Name'].str.rstrip('_')
```

```python
df
```

## Cleaning, standardizing phone numbers

```python
# Show df
df
```

```python
# Clean the phone numbers by removing non-alphanumeric characters  
df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)  
```

```python
# Clean the phone numbers by removing non-alphanumeric characters  
df["Phone_Number"] = df["Phone_Number"].str.replace('[^a-zA-Z0-9]', '', regex=True)

# Show the dataframe
df
```

```python
# Convert all entries in the 'Phone_Number' column to strings  
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
# Apply the changes to the column 'Phone_Number'
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: str(x))
```

```python
# Format the 'Phone_Number' column by adding dashes in the format XXX-XXX-XXXX  
df['Phone_Number'] = df['Phone_Number'].apply(lambda x: x[0:3] + '-' + x[3:6] + '-' + x[6:10])
```

```python
# Show df
df
```

```python
# Remove the substring 'nan--' from the 'Phone_Number' column  
df["Phone_Number"] = df["Phone_Number"].str.replace('nan--', '', regex=True)
```

```python
# Remove the substring 'Na--' from the 'Phone_Number' column  
df["Phone_Number"] = df["Phone_Number"].str.replace('Na--', '', regex=True)
```

```python
# Show df
df
```

## Splitting columns

```python
# Show the column "Address"
df["Address"]
```

```python
# Split the 'Address' column by commas and expand the results into separate columns
df["Address"].str.split(',', expand = True)
```

```python
# Apply the changes and name the columns 'Street_Adress', 'State' & 'Zip_Code'
df[['Street_Adress', 'State', 'Zip_Code']] = df["Address"].str.split(',', expand = True)
```

```python
# Show df
df
```

## Standardizing column values using replace

```python
# Replace the values 'Yes' with 'Y' in the 'Paying Customer' column  
df["Paying Customer"].str.replace('Yes', 'Y', regex=True)
```

```python
# Aplly the changes
df["Paying Customer"] = df["Paying Customer"].str.replace('Yes', 'Y', regex=True)
```

```python
# Replace the values 'No' with 'N' in the 'Paying Customer' column 
df["Paying Customer"] = df["Paying Customer"].str.replace('No', 'N', regex=True)
```

```python
# Show df
df
```

```python
# Replace the values 'Yes' with 'Y' in the 'Do_Not_Contact' column and apply the changes
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('Yes', 'Y', regex=True)

# Replace the values 'No' with 'N' in the 'Do_Not_Contact' column and apply the changes
df["Do_Not_Contact"] = df["Do_Not_Contact"].str.replace('No', 'N', regex=True)

# Show the column "Do_Not_Contact"
df["Do_Not_Contact"]
```

## Fill null values

```python
# Show df
df
```

```python
# Replace occurrences of 'N/a' with an empty string in the entire DataFrame  
df = df.replace('N/a', '', regex=True)
```

```python
# Apply the changes
df = df.replace('N/a', '', regex = True)
```

```python
# Fill NaN values in the entire DataFrame with an empty string  
df = df.fillna('')
```

```python
# Apply the changes
df = df.fillna('')
```

```python
# Show df
df
```

## Filtering down rows of data

- The list of the phone numbers we can call:

```python
# Loop through the DataFrame index and drop rows where 'Do_Not_Contact' is 'Y'  
for x in df.index:  
    if df.loc[x, 'Do_Not_Contact'] == 'Y':  
        df.drop(x, inplace=True)

# Show df
df
```

```python
# Loop through the DataFrame index and drop rows where 'Phone_Number' is empty 
for x in df.index:
    if df.loc[x, 'Phone_Number'] == '':
        df.drop(x, inplace = True)

# Show df
df
```

```python
# Reset the index of the dataFrame to avoid the gaps showing
df = df.reset_index(drop = True)

# Show df
df
```
